```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

class LinearRegression():
    def __init__(self, X, y, initial_guess, alpha=0.99):
        self.X = np.append(np.ones((m,1)),X, axis=1)
        self.y = y
        self.theta = np.append(np.array([1]),initial_guess)
        
        self.m = len(y)
        self.alpha = alpha
        
    def _new_theta(self):
        change = (self.alpha/self.m)* np.matmul( \
                    (self.model()-self.y),self.X )
        return self.theta - change
    
    def optimize_grad(self, record_J=False):
        J_list = None
        if record_J: J_list = []
        for i in range(1000):
            self.theta = self._new_theta() 
            # make into while loop. Calculate J gradient. When stable, break
            
            # for snapshoting J
            if record_J:
                J_list.append(cost_J(self.X, self.y, self.theta))
        return self.theta, J_list
        
    def model(self):
        X_T = np.transpose(self.X)
        return np.matmul(self.theta,X_T)
    
    def cost_J(self):
        return np.sum(
                    np.square(self.model(self.X,self.theta)-self.y)
                    )/(2*m)
```
